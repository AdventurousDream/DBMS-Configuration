```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

## Preprocess

```python
# extract data
raw_data = pd.read_csv("./data.csv")
data = raw_data
data.tail()
```

```python
col4epoch = data.columns[[1, 2, 3, 4, 6]]
col4epoch
```

## Plot

```python
data0 = data[data['Episode'] == 0][col4epoch]
g = sns.PairGrid(data = data0)
g = g.map(plt.scatter, s = 10)
g.savefig('epoch01.png', dpi=150)
```

```python
data1 = data[data['Episode'] == 1][col4epoch]
g = sns.PairGrid(data = data1)
g = g.map(plt.scatter, s = 10)
g.savefig('epoch02.png', dpi=150)
```
